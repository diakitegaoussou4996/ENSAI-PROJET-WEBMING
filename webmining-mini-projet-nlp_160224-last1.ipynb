<center><h1>Lecture des donn√©es du flux RSS</h1></center>

<p style="font-family: 'Arial', sans-serif; font-size: 14px;">
    Voici une explication d√©taill√©e de la d√©marche que nous avons suivie pour analyser et extraire les donn√©es du flux RSS :
</p>

<ol style="font-family: 'Arial', sans-serif; font-size: 14px;">
    <li><strong>Lien vers le fichier tvinfo-sources.json :</strong> Nous avons d'abord d√©fini le lien vers le fichier tvinfo-sources.json, qui contient les URL des flux RSS que nous voulons analyser.</li>
    <li><strong>Charger la liste des flux RSS :</strong> Nous avons utilis√© la biblioth√®que requests pour r√©cup√©rer le contenu du fichier tvinfo-sources.json depuis son URL. Ensuite, nous avons converti le contenu JSON en une liste de sources.</li>
    <li><strong>Lecture des donn√©es du flux RSS :</strong> Nous avons parcouru la liste des sources et pour chaque source, nous avons utilis√© la biblioth√®que feedparser pour lire les donn√©es du flux RSS correspondant.</li>
    <li><strong>R√©cup√©ration du premier article :</strong> Pour chaque flux RSS, nous avons r√©cup√©r√© le premier article s'il y en avait un.</li>
    <li><strong>Extraction des informations de l'article :</strong> Nous avons extrait diff√©rentes informations de l'article telles que le titre, la description, la date de publication, le lien, la cat√©gorie, le contenu et le lien de l'image (si disponibles).</li>
    <li><strong>Affichage des informations de l'article :</strong> Enfin, nous avons affich√© toutes les informations extraites de l'article.</li>
</ol>

<p style="font-family: 'Arial', sans-serif; font-size: 14px;">
    Cette d√©marche nous permet d'obtenir un aper√ßu des derniers articles de chaque source de flux RSS, ce qui est utile pour une analyse plus approfondie des informations.
</p>


```python
# Lien vers le fichier tvinfo-sources.json
tvinfo_sources_url = "https://people.irisa.fr/Guillaume.Gravier/teaching/ENSAI/data/tvinfo-sources.json"

# Charger la liste des flux RSS depuis le lien tvinfo-sources.json
response_sources = requests.get(tvinfo_sources_url)
sources = json.loads(response_sources.text)

print(f'-------------------------------------------------Les flux RSS qui sont dans le fichier tvinfo-sources.json sont-------------------------------------------------')

print(sources)


print(f'-------------------------------------------------------Lecture des donn√©es du flux RSS-------------------------------------------------------------------------')

# Parcourir les flux RSS
for source_name, source_url in sources.items():
    # Lire les donn√©es du flux RSS
    data = fp.parse(source_url)
    
    # R√©cup√©rer le premier article du flux RSS (s'il y en a)
    if data.entries:
        item = data.entries[0]  # Prendre le premier article
        
        # Extraire les informations de l'article
        article = {
            "source": source_name,
            "title": item.title,
            "description": item.description,
            "pubDate": item.published,
            "link": item.link,
            "category": source_name,  # Vous pouvez modifier cette valeur en fonction de la source si n√©cessaire
            "content": item.content[0].value if "content" in item else "",  # Si le champ "content" est disponible
            "image_link": item.enclosures[0].url if "enclosures" in item and item.enclosures[0].type == "image/jpeg" else ""  # Si le champ "enclosures" est disponible et de type image/jpeg
        }
        
        # Afficher les informations de l'article
        print("Source:", article["source"])
        print("Titre:", article["title"])
        print("Description:", article["description"])
        print("Date de publication:", article["pubDate"])
        print("Lien:", article["link"])
        print("Cat√©gorie:", article["category"])
        print("Contenu:", article["content"])
        print("Lien de l'image:", article["image_link"])
        print("\n")
```

<center><h1>Scanner les Articles et Rep√©rer les Nouveaux</h1></center>
<hr style="border-width:6px;border-color:#75DFC1">

<p style="font-family: 'Arial', sans-serif; font-size: 14px;">
    Apr√®s avoir obtenu les derniers articles de chaque source de flux RSS, nous avons proc√©d√© √† la t√¢che de scanner les articles pour rep√©rer ceux qui ne sont pas d√©j√† pr√©sents dans la base de donn√©es. Voici comment nous avons accompli cette √©tape :
</p>

<ol style="font-family: 'Arial', sans-serif; font-size: 14px;">
    <li><strong>Lien vers la base de donn√©es existante :</strong> Nous avons d√©fini le lien vers la base de donn√©es existante, qui √©tait stock√©e dans le fichier francetvinfo.json.</li>
    <li><strong>Charger la base de donn√©es existante :</strong> Nous avons utilis√© la biblioth√®que requests pour r√©cup√©rer le contenu du fichier francetvinfo.json depuis son URL. Ensuite, nous avons converti le contenu JSON en une base de donn√©es Python.</li>
    <li><strong>Liste des nouveaux articles :</strong> Nous avons cr√©√© une liste vide appel√©e "new_articles" pour stocker les nouveaux articles que nous allons rep√©rer.</li>
    <li><strong>Parcourir les flux RSS :</strong> Nous avons parcouru √† nouveau les flux RSS, en utilisant la biblioth√®que feedparser pour lire les donn√©es de chaque flux.</li>
    <li><strong>Parcourir les articles du flux RSS :</strong> Pour chaque flux RSS, nous avons parcouru les articles en utilisant une boucle.</li>
    <li><strong>Extraction des informations de l'article :</strong> Nous avons extrait certaines informations cl√©s de chaque article, telles que le titre, la date de publication et le lien.</li>
    <li><strong>V√©rification dans la base de donn√©es :</strong> Nous avons v√©rifi√© si l'article existe d√©j√† dans la base de donn√©es en utilisant l'URL de l'article comme cl√© unique. Si l'article n'√©tait pas d√©j√† pr√©sent, nous l'avons ajout√© √† la liste des nouveaux articles.</li>
</ol>

<p style="font-family: 'Arial', sans-serif; font-size: 14px;">
    √Ä la fin de cette √©tape, la liste "new_articles" contient les nouveaux articles qui ne sont pas d√©j√† pr√©sents dans la base de donn√©es. Ces articles peuvent ensuite √™tre examin√©s plus en d√©tail ou trait√©s selon les besoins.

</p>

<h2 style="font-family: 'Arial', sans-serif; font-size: 16px;">D√©marche du Code</h2>
<p style="font-family: 'Arial', sans-serif; font-size: 14px;">
    Le code ci-dessous d√©taille les √©tapes pour accomplir cette t√¢che :
</p>

<ol style="font-family: 'Arial', sans-serif; font-size: 14px;">
    <li><strong>Liste des Nouveaux Articles :</strong> Nous commen√ßons par cr√©er une liste vide pour stocker les nouveaux articles.</li>
    <li><strong>Parcourir les Flux RSS :</strong> Nous parcourons les flux RSS sp√©cifi√©s pour obtenir les informations des articles.</li>
    <li><strong>Parcourir les Articles du Flux RSS :</strong> Pour chaque article dans le flux RSS, nous collectons des informations telles que le titre, la date de publication, le lien, la description, le contenu (initialis√© √† vide), et le lien de l'image (s'il y en a).</li>
    <li><strong>V√©rification de l'Existence :</strong> Nous v√©rifions si l'article existe d√©j√† dans la base de donn√©es en utilisant l'URL comme cl√© unique.</li>
    <li><strong>Mise √† Jour des Informations de l'Article :</strong> Si l'article est nouveau, nous cr√©ons un objet Article √† partir de l'URL de l'article, t√©l√©chargeons la page web, et analysons la page web pour extraire le contenu textuel complet de l'article, les auteurs, et l'URL de l'image.</li>
    <li><strong>Ajout √† la Liste des Nouveaux Articles :</strong> Nous ajoutons ensuite l'article √† la liste des nouveaux articles.</li>
    <li><strong>Mise √† Jour de la Base de Donn√©es :</strong> Apr√®s avoir parcouru tous les articles, nous mettons √† jour la base de donn√©es existante avec les nouveaux articles.</li>
</ol>

<p style="font-family: 'Arial', sans-serif; font-size: 14px;">
    Apr√®s l'ex√©cution du code, la base de donn√©es sera mise √† jour avec le contenu textuel complet des articles ainsi que les noms des auteurs lorsque disponibles. Elle sera pr√™te √† √™tre utilis√©e pour un traitement ult√©rieur.
</p>
</body>
</html>


```python
def mettre_a_jour_base_de_donnees(flux_rss_a_lire, database):
    # Liste des nouveaux articles
    new_articles = []

    # Parcourir les flux RSS
    for source_url in flux_rss_a_lire:
        data = fp.parse(source_url)
        
        # Parcourir les articles du flux RSS
        for item in data.entries:
            article = {
                "title": item.title,
                "published": item.published,
                "link": item.link,
                "description": item.description if "description" in item else "",
                "content": "",  # Initialiser le contenu √† vide
                "image_link": item.enclosures[0].url if "enclosures" in item and item.enclosures[0].type == "image/jpeg" else ""
            }
            
            # V√©rifier si l'article existe d√©j√† dans la base de donn√©es en utilisant l'URL comme cl√© unique
            if article["link"] not in database:
                # Cr√©er un objet Article √† partir de l'URL de l'article
                article_obj = np.Article(article["link"])
                article_obj.download()
                article_obj.parse()
                
                # Mettre √† jour les informations de l'article
                article["content"] = article_obj.text
                article["authors"] = article_obj.authors
                article["image_url"] = article_obj.top_image
                
                # Ajouter l'article √† la liste des nouveaux articles
                new_articles.append(article)

    # Mettre √† jour la base de donn√©es existante avec les nouveaux articles
    for article in new_articles:
        database[article["link"]] = article

    # Retourner la base de donn√©es mise √† jour
    return database

# Liste des flux RSS √† parcourir
flux_rss_a_lire = list(sources.values())

# Charger la base de donn√©es existante depuis le lien francetvinfo.json
francetvinfo_url = "https://people.irisa.fr/Guillaume.Gravier/teaching/ENSAI/data/francetvinfo.json"
response_database = requests.get(francetvinfo_url)
database = json.loads(response_database.text)

# Mettre √† jour la base de donn√©es avec les nouveaux articles
database_mise_a_jour = mettre_a_jour_base_de_donnees(flux_rss_a_lire, database)

# Maintenant, database_mise_a_jour contient la base de donn√©es mise √† jour avec le contenu et les auteurs des articles en plus des autres informations.
```

```python
# Fonction pour extraire les verbes entre deux entit√©s dans la m√™me phrase
def extract_verbs_between_entities(entity1, entity2, entity_type1, entity_type2):
    # Initialiser une liste pour stocker les phrases contenant les deux entit√©s
    sentences_with_entities = []
    
    # Parcourir les articles et leurs entit√©s
    for article_url, article_entities in articles_entities.items():
        # V√©rifier si les deux entit√©s de types correspondants apparaissent dans la m√™me phrase
        if any(entity["text"] == entity1 and entity["type"] == entity_type1 for entity in article_entities) \
            and any(entity["text"] == entity2 and entity["type"] == entity_type2 for entity in article_entities):
            # Extraire la phrase contenant les entit√©s
            article_content = database[article_url]['content']
            doc = nlp(article_content)
            for sentence in doc.sents:
                sentence_entities = [entity["text"] for entity in article_entities if entity["text"] in sentence.text]
                if entity1 in sentence_entities and entity2 in sentence_entities:
                    sentences_with_entities.append(sentence.text)
    
    # Traiter les phrases avec spaCy pour extraire les verbes entre les entit√©s
    verbs_between = []
    for sentence in sentences_with_entities:
        doc = nlp(sentence)
        for token in doc:
            if token.pos_ == "VERB":
                verbs_between.append(token.lemma_)
    
    return verbs_between
```
