Chaque article est analys√© pour en extraire le titre, la date de publication, le r√©sum√©, le lien, et d'autres informations disponibles comme le lien de l'image et le contenu suppl√©mentaire.</li>

```python
# URL du flux RSS √† analyser
url = "https://www.lemonde.fr/rss/une.xml"

# R√©cup√©ration et analyse du flux RSS
data = fp.parse(url)

# Afficher les informations du flux RSS
print("Titre du flux RSS:", data.feed.title)
print("Date de publication du flux RSS:", data.feed.published)

# It√©ration sur quelques articles du flux RSS
for item in data.entries[:3]:  # Limite √† trois articles pour l'exemple
    print("\nArticle:")
    print("Titre:", item.title)
    print("Date de publication:", item.published)
    print("R√©sum√©:", item.summary)
    print("Lien:", item.link)
    if 'media_content' in item:
        print("Lien de l'image:", item.media_content[0]['url'])
    if 'content' in item:
        print("Contenu suppl√©mentaire:", item.content[0]['value'])
```

    Voici une explication d√©taill√©e de la d√©marche que nous avons suivie pour analyser et extraire les donn√©es du flux RSS :
</p>

<ol style="font-family: 'Arial', sans-serif; font-size: 14px;">
    <li><strong>Lien vers le fichier tvinfo-sources.json :</strong> Nous avons d'abord d√©fini le lien vers le fichier tvinfo-sources.json, qui contient les URL des flux RSS que nous voulons analyser.</li>
    <li><strong>Charger la liste des flux RSS :</strong> Nous avons utilis√© la biblioth√®que requests pour r√©cup√©rer le contenu du fichier tvinfo-sources.json depuis son URL. Ensuite, nous avons converti le contenu JSON en une liste de sources.</li>
    <li><strong>Lecture des donn√©es du flux RSS :</strong> Nous avons parcouru la liste des sources et pour chaque source, nous avons utilis√© la biblioth√®que feedparser pour lire les donn√©es du flux RSS correspondant.</li>
    <li><strong>R√©cup√©ration du premier article :</strong> Pour chaque flux RSS, nous avons r√©cup√©r√© le premier article s'il y en avait un.</li>
    <li><strong>Extraction des informations de l'article :</strong> Nous avons extrait diff√©rentes informations de l'article telles que le titre, la description, la date de publication, le lien, la cat√©gorie, le contenu et le lien de l'image (si disponibles).</li>
    <li><strong>Affichage des informations de l'article :</strong> Enfin, nous avons affich√© toutes les informations extraites de l'article.</li>
</ol>

<p style="font-family: 'Arial', sans-serif; font-size: 14px;">
    Cette d√©marche nous permet d'obtenir un aper√ßu des derniers articles de chaque source de flux RSS, ce qui est utile pour une analyse plus approfondie des informations.
</p>


```python
# Lien vers le fichier tvinfo-sources.json
tvinfo_sources_url = "https://people.irisa.fr/Guillaume.Gravier/teaching/ENSAI/data/tvinfo-sources.json"

# Charger la liste des flux RSS depuis le lien tvinfo-sources.json
response_sources = requests.get(tvinfo_sources_url)
sources = json.loads(response_sources.text)

print(f'-------------------------------------------------Les flux RSS qui sont dans le fichier tvinfo-sources.json sont-------------------------------------------------')

print(sources)


print(f'-------------------------------------------------------Lecture des donn√©es du flux RSS-------------------------------------------------------------------------')

# Parcourir les flux RSS
for source_name, source_url in sources.items():
    # Lire les donn√©es du flux RSS
    data = fp.parse(source_url)
    
    # R√©cup√©rer le premier article du flux RSS (s'il y en a)
    if data.entries:
        item = data.entries[0]  # Prendre le premier article
        
        # Extraire les informations de l'article
        article = {
            "source": source_name,
            "title": item.title,
            "description": item.description,
            "pubDate": item.published,
            "link": item.link,
            "category": source_name,  # Vous pouvez modifier cette valeur en fonction de la source si n√©cessaire
            "content": item.content[0].value if "content" in item else "",  # Si le champ "content" est disponible
            "image_link": item.enclosures[0].url if "enclosures" in item and item.enclosures[0].type == "image/jpeg" else ""  # Si le champ "enclosures" est disponible et de type image/jpeg
        }
        
        # Afficher les informations de l'article
        print("Source:", article["source"])
        print("Titre:", article["title"])
        print("Description:", article["description"])
        print("Date de publication:", article["pubDate"])
        print("Lien:", article["link"])
        print("Cat√©gorie:", article["category"])
        print("Contenu:", article["content"])
        print("Lien de l'image:", article["image_link"])
        print("\n")
```

```python
def mettre_a_jour_base_de_donnees(flux_rss_a_lire, database, categories):
    # Liste des nouveaux articles
    new_articles = []

    # Parcourir les flux RSS
    for source_url in flux_rss_a_lire:
        # D√©terminer la cat√©gorie du flux
        category = categories[source_url]

        data = fp.parse(source_url)
        
        # Parcourir les articles du flux RSS
        for item in data.entries:
            # Cl√©s renomm√©es pour correspondre √† la base de donn√©es existante
            article = {
                "title": item.title,
                "date": item.published,  # Renomm√© de 'published' √† 'date'
                "url": item.link,  # Renomm√© de 'link' √† 'url'
                "content": item.description if "description" in item else "",
                "image_link": item.enclosures[0].url if "enclosures" in item and item.enclosures[0].type == "image/jpeg" else "",
                "category": category,  # Ajout√© la cat√©gorie
            }
            
            # V√©rifier si l'article existe d√©j√† dans la base de donn√©es
            if article["url"] not in database:
                article_obj = np.Article(article["url"])
                article_obj.download()
                article_obj.parse()
                
                # Mettre √† jour les informations de l'article
                article["content"] += "\n\n" + article_obj.text  # Ajouter le contenu t√©l√©charg√©
                article["author"] = article_obj.authors  # Ajout√© la liste des auteurs
                
                # Ajouter l'article √† la liste des nouveaux articles
                new_articles.append((article["url"], article))

    # Mettre √† jour la base de donn√©es existante avec les nouveaux articles
    for url, article in new_articles:
        database[url] = article

    return database


# Liste des flux RSS √† parcourir
flux_rss_a_lire = list(sources.values())

# Usage de la fonction avec les cat√©gories
categories = {
    "https://www.francetvinfo.fr/france.rss": "france",
    "https://www.francetvinfo.fr/monde/europe.rss": "europe",  # Corrig√© ici
    "https://www.francetvinfo.fr/economie/entreprises.rss": "economie"
}


# Charger la base de donn√©es existante depuis le lien francetvinfo.json
francetvinfo_url = "https://people.irisa.fr/Guillaume.Gravier/teaching/ENSAI/data/francetvinfo.json"
response_database = requests.get(francetvinfo_url)
database = json.loads(response_database.text)

# Mettre √† jour la base de donn√©es avec les nouveaux articles
database_mise_a_jour = mettre_a_jour_base_de_donnees(flux_rss_a_lire, database, categories)

# Mettre √† jour la base de donn√©es avec les nouveaux articles
database_mise_a_jour = mettre_a_jour_base_de_donnees(flux_rss_a_lire, database, categories)

# Transformer la base de donn√©es mise √† jour en DataFrame
df_database_mise_a_jour = pd.DataFrame(database_mise_a_jour.values())

# Ajouter les URLs comme colonne dans le DataFrame
df_database_mise_a_jour['url'] = list(database_mise_a_jour.keys())

# Afficher les premi√®res lignes pour v√©rifier
df_database_mise_a_jour

```

Cette fonction prend en entr√©e une paire d'entit√©s nomm√©es ainsi que leurs types correspondants (par exemple, une personne et une organisation) et calcule le nombre de co-occurrences de ces deux entit√©s au sein d'un m√™me document de notre collection.
</p>

<p style="font-family: 'Arial', sans-serif; font-size: 14px;">
    La d√©tection des co-occurrences est une technique couramment utilis√©e pour identifier des relations potentielles entre des entit√©s. Une fr√©quence de co-occurrence √©lev√©e sugg√®re g√©n√©ralement un lien significatif entre les entit√©s, ce qui peut √™tre utile pour des analyses ult√©rieures telles que l'identification de partenariats, de citations ou de th√®mes r√©currents.

<p style="font-family: 'Arial', sans-serif; font-size: 14px;">

```python
def count_entity_cooccurrences(entity1, entity2, entity_type1, entity_type2, articles_entities):
    cooccurrence_count = 0

    """
    Cette fonction compte le nombre de co-occurrences de deux entit√©s d'un m√™me type dans la collection de documents.
    
    Args:
        entity1 (str): Le texte de la premi√®re entit√©.
        entity2 (str): Le texte de la deuxi√®me entit√©.
        type1 (str): Le type de la premi√®re entit√© (PER, ORG ou LOC).
        type2 (str): Le type de la deuxi√®me entit√© (PER, ORG ou LOC).
        articles_entities (dict): Un dictionnaire contenant les entit√©s extraites pour chaque document.

    Returns:
        int: Le nombre de co-occurrences de ces deux entit√©s dans la collection.
    """
    # Parcourir les articles et leurs entit√©s
    for entities_info in articles_entities.values():
        # Filtrer les entit√©s par type et texte
        entities_of_type1 = [entity for entity in entities_info if entity["type"] == entity_type1 and entity["text"] == entity1]
        entities_of_type2 = [entity for entity in entities_info if entity["type"] == entity_type2 and entity["text"] == entity2]
        
        # Compter la co-occurrence si les deux types d'entit√©s sp√©cifiques sont pr√©sents dans le m√™me article
        if entities_of_type1 and entities_of_type2:
            cooccurrence_count += 1
    
    return cooccurrence_count
# Exemple d'utilisation de la fonction
entity1 = "France"
entity2 = "Covid-19"
entity_type1 = "LOC"
entity_type2 = "PER"
# Appel de la fonction pour compter les co-occurrences
cooccurrence_count = count_entity_cooccurrences(entity1, entity2, entity_type1, entity_type2, articles_entities)

# Affichage du r√©sultat
print(f"Le nombre de co-occurrences de '{entity1}' ({entity_type1}) et '{entity2}' ({entity_type2}) est de {cooccurrence_count}")
```

```python
# Cr√©er une fonction pour rechercher les paires d'entit√©s de types diff√©rents qui apparaissent le plus souvent ensemble
def find_most_frequent_entity_pairs(articles_entities):
    entity_pairs_count = {}  # Dictionnaire pour stocker les comptes des paires d'entit√©s
    
    # Parcourir les articles et leurs entit√©s
    for entities_info in articles_entities.values():
        # Cr√©er des ensembles distincts d'entit√©s pour chaque type
        people_entities = set()
        organization_entities = set()
        location_entities = set()
        
        for entity_info in entities_info:
            entity_text = entity_info["text"]
            entity_type = entity_info["type"]
            
            # Ajouter l'entit√© √† l'ensemble correspondant
            if entity_type == "PER":
                people_entities.add(entity_text)
            elif entity_type == "ORG":
                organization_entities.add(entity_text)
            elif entity_type == "LOC":
                location_entities.add(entity_text)
        
        # G√©n√©rer des paires de types diff√©rents et les compter
        for entity1 in people_entities:
            for entity2 in organization_entities.union(location_entities):
                entity_pair = (entity1, entity2)
                entity_pairs_count[entity_pair] = entity_pairs_count.get(entity_pair, 0) + 1
    
    # Trier les paires par fr√©quence d√©croissante
    sorted_entity_pairs = sorted(entity_pairs_count.items(), key=lambda x: x[1], reverse=True)
    
    return sorted_entity_pairs

# Appel de la fonction pour trouver les paires d'entit√©s
frequent_entity_pairs = find_most_frequent_entity_pairs(articles_entities)

# Affichage des 10 paires d'entit√©s les plus fr√©quentes
top_10_entity_pairs = frequent_entity_pairs[:10]
for pair, count in top_10_entity_pairs:
    entity1, entity2 = pair
    print(f"Paire d'entit√©s : '{entity1}' et '{entity2}' - Fr√©quence : {count}")
```

```python
# Charger le mod√®le spaCy
nlp = spacy.load("fr_core_news_md", disable=["parser", "tagger"])
nlp.enable_pipe("senter")  # d√©coupage en phrases est activ√© ? 

# Articles_entities contient les articles avec leur contenu
articles_contents = [article['content'] for article in database.values()]

# Initialiser un dictionnaire pour compter les co-occurrences
co_occurrences = defaultdict(int)

# Traiter les textes par lots avec nlp.pipe
for doc in nlp.pipe(articles_contents, batch_size=50):
    for sent in doc.sents:  # Parcourir les phrases du document
        # Trouver les paires d'entit√©s de types diff√©rents dans la m√™me phrase
        entities = list(sent.ents)  # Convertir en liste pour l'it√©ration
        for i, entity1 in enumerate(entities):
            for entity2 in entities[i+1:]:  # Comparer avec les entit√©s suivantes
                if entity1.label_ != entity2.label_:  # S'assurer que les types sont diff√©rents
                    # Cr√©er une paire tri√©e pour √©viter les doublons (entit√©1, entit√©2) == (entit√©2, entit√©1)
                    pair = tuple(sorted([entity1.text, entity2.text]))
                    co_occurrences[pair] += 1

# Trier les co-occurrences par fr√©quence d√©croissante
sorted_co_occurrences = sorted(co_occurrences.items(), key=lambda x: x[1], reverse=True)

# Afficher les 15 paires les plus fr√©quentes
for pair, freq in sorted_co_occurrences[:15]:
    print(f"{pair}: {freq}")
```

```python
# Fonction pour extraire les verbes entre deux entit√©s dans la m√™me phrase
def extract_verbs_between_entities(entity1, entity2, entity_type1, entity_type2):
    # Initialiser une liste pour stocker les phrases contenant les deux entit√©s
    sentences_with_entities = []
    
    # Parcourir les articles et leurs entit√©s
    for article_url, article_entities in articles_entities.items():
        # V√©rifier si les deux entit√©s de types correspondants apparaissent dans la m√™me phrase
        if any(entity["text"] == entity1 and entity["type"] == entity_type1 for entity in article_entities) \
            and any(entity["text"] == entity2 and entity["type"] == entity_type2 for entity in article_entities):
            # Extraire la phrase contenant les entit√©s
            article_content = database[article_url]['content']
            doc = nlp(article_content)
            for sentence in doc.sents:
                sentence_entities = [entity["text"] for entity in article_entities if entity["text"] in sentence.text]
                if entity1 in sentence_entities and entity2 in sentence_entities:
                    sentences_with_entities.append(sentence.text)
    
    # Traiter les phrases avec spaCy pour extraire les verbes entre les entit√©s
    verbs_between = []
    for sentence in sentences_with_entities:
        doc = nlp(sentence)
        for token in doc:
            if token.pos_ == "VERB":
                verbs_between.append(token.lemma_)
    
    return verbs_between
```
